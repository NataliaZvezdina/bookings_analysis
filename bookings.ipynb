Analyze hotel booking data. Description:
*    `Hotel` - type of hotel (City Hotel or Resort Hotel)
*    `Is canceled` - booking has been canceled (1) or not (0); not canceled is considered successful
*    `Lead time` - number of days between booking date and arrival date
*    `Arrival full date` - full arrival date
*    `Arrival date year` - year of arrival
*    `Arrival date month` - month of arrival
*    `Arrival date week number` - week of arrival
*    `Arrival date day of month` - day of arrival
*    `Stays in weekend nights` - the number of weekends (Saturday or Sunday) that the guest has booked for hotel stay
*    `Stays in week nights` - the number of days (Monday to Friday) that the guest has booked for stay at the hotel
*    `Stays total nights` - total number of nights booked (sum of two previous columns)
*    `Adults` - number of adults
*    `Children` - number of children
*    `Babies` - the number of babies
*    `Meal` - selected type of food
*    `Country` - the country of origin of the client
*    `Reserved room type` - type of reserved room
*    `Assigned room type` - the type of room received (may differ from the booked)
*    `Customer type` - type of booking
*    `Reservation status` - the value of the last reservation status: Canceled - was canceled by the client; Check-Out - the client checked in but has already left the hotel; No-Show - the client did not register and told the hotel the reason
*    `Reservation status date` - date of status update

Analyze the dates of scheduled arrival:
* Which month were the largest number of successful bookings made in 2016? Has the most popular month in 2017 changed?
* Group by year and check which month the City Hotel type bookings were canceled for most times during each period?

Create a `total_kids` column by combining children and babies. What type of hotels on average are more popular with clients with children?

```python
df['total_kids'] = df['children'] + df['babies']
df.total_kids
```

```python
df \
    .groupby('hotel') \
    .aggregate({'total_kids': 'mean'})
```

As we can see, `Resort Hotel` is the most popular among guests with kids.

***
Create a `has_kids` variable that accepts the `True` value if the client specifies `at least one child` (total_kids) when booking, otherwise - `False`. 
Calculate the ratio of the number of retired users to the total number of clients, expressed as a percentage (`churn rate`). Indicate which group has the highest score.

```python
df['has_kids'] = df['total_kids'] > 0
```

```python
df[['has_kids']].sample(n=10)
```

```python
retire_clients_has_kids_num = df.query('has_kids == True and is_canceled == 1').shape[0]
total_clients_has_kids_num = df.query('has_kids == True').shape[0]

print(f'{retire_clients_has_kids_num=}')
print(f'{total_clients_has_kids_num=}')

churn_rate_has_kids = round(retire_clients_has_kids_num / total_clients_has_kids_num * 100, 2)
print(f'{churn_rate_has_kids=}')
```

```python
retire_clients_no_kids_num = df.query('has_kids == False and is_canceled == 1').shape[0]
total_clients_no_kids_num = df.query('has_kids == False').shape[0]

print(f'{retire_clients_no_kids_num=}')
print(f'{total_clients_no_kids_num=}')

churn_rate_no_kids = round(retire_clients_no_kids_num / total_clients_no_kids_num * 100, 2)
print(f'{churn_rate_no_kids=}')
```

As calculations show, `churn rate` among clients that have no kids - `37.22` is more than one among clients with kids - `34.92`.
